```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV exported from your dashboard
df = pd.read_csv("variantC_run.csv")

# ---- Compute metrics ----
Vrms = np.sqrt(np.mean(df['v_V']**2))
Vpeak = df['v_V'].abs().max()
Irms = np.sqrt(np.mean(df['i_A']**2))
Ipeak = df['i_A'].abs().max()
Pavg = df['p_W'].mean()

print("Vrms = {:.2f} V".format(Vrms))
print("Vpeak = {:.2f} V".format(Vpeak))
print("Irms = {:.3f} A".format(Irms))
print("Ipeak = {:.3f} A".format(Ipeak))
print("Pavg = {:.3f} W".format(Pavg))

# ---- Plot signals ----
plt.figure(figsize=(10,6))

# Voltage and current
plt.subplot(2,1,1)
plt.plot(df['time_s'], df['v_V'], label="v(t) [V]")
plt.plot(df['time_s'], df['i_A'], label="i(t) [A]")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

# Power
plt.subplot(2,1,2)
plt.plot(df['time_s'], df['p_W'], label="p(t) [W]", color="purple")
plt.xlabel("Time [s]")
plt.ylabel("Power [W]")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
